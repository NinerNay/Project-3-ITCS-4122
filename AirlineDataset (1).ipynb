##Visualizations

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.barplot(data=airline_df, x = 'carrier_name', y = 'arr_delay')
plt.xticks(rotation=90)
plt.show()
#Displays the airlines with the highest amount of arrival delays
```

Some airline companies consistently outperform their competitors. In the visualization we can see that Southwest Airlines and  American Airlines had the most amount of delays. This can be common when dealing with legacy carriers due to operating far more flights than the smaller carriers. Furthermore, although Spirit airlines arenâ€™t considered a legacy airline, they still contribute significantly to the amount of delays. High totals in this case doesnâ€™t automatically mean that the airline is necessarily a â€œbad airlineâ€.

```python
# Creates a date column
new_df['date'] = pd.to_datetime(new_df[['year','month']].assign(day=1))

# Grouping by date... includes all the delays
delay_trends = new_df.groupby('date')[['carrier_delay','weather_delay','nas_delay','security_delay','late_aircraft_delay']].sum()

# Actual plot displaying the different delay trends over the course of 10 years
plt.figure(figsize=(14,6))
for col in delay_trends.columns:
    plt.plot(delay_trends.index, delay_trends[col], label=col)

plt.title("Delay Trends Over Time (2013â€“2023)", fontsize=14)
plt.xlabel("Year")
plt.ylabel("Total Delay Minutes")
plt.legend(title="Cause of Delay")
plt.show()
```

This chart helps visualize how different types of delays have changed over the 10-year period. Many of the weather-related delays tend to spike seasonally, where factors like snow, or hurricane season can affect these decisions. Furthermore, NAS (National Airspace System) delays follow a steadier trend staying consistent over the years. Carrier and late-aircraft delays also show recurring patterns, suggesting operational issues that persist over time. This chart gives us insight into whether delays are improving, worsening as years go by.

```python
values = [
    new_df['arr_del15'].sum(),
    new_df['arr_cancelled'].sum(),
    new_df['arr_diverted'].sum()
]
#added so it's easier to see what each value is
labels = ["Delays (15+ min)", "Cancellations", "Diversions"]

# Donut chart
#used ai to help determine how to show percentage
plt.pie(values, labels=labels, autopct='%1.1f%%', startangle=10, wedgeprops=dict(width=0.4))
plt.title("Distribution of Flight Disruptions")
plt.show()
```


This chart gives us the big picture of flight disruptions within the years. The majority of disruptions are delays, while cancellations and diversions make up a smaller portion. While cancellations and diversions may be less frequent, their impact on passengers is often more severe. This can often leave the customer stranded in an unknown state, with little to no compensation from the airlines. This comparison shows that most problems passengers face come from waiting, rather than outright losing their flight, but it also highlights the need to address cancellation risks.

```python
# Group by airlines... also by the sum of the delay causes
delay_causes = new_df.groupby('carrier_name')[['carrier_delay','weather_delay','nas_delay','security_delay','late_aircraft_delay']].sum()

# Plot stacked bar shows a breakdown on the delay causes by the airline
delay_causes.plot(kind='bar', stacked=True, figsize=(12,6))#fig size to make it more readable
plt.title("Breakdown of Delay Causes by Airline")
plt.xlabel("Airline")
plt.ylabel("Total Delay Minutes")
plt.legend(title="Cause of Delay")
plt.show()

```

```python


# Group by airport and sum the delay causes
delay_causes_airport = new_df.groupby('airport')[['carrier_delay','weather_delay','nas_delay','security_delay','late_aircraft_delay']].sum()

# Calculate total delays per airport
delay_causes_airport['total_delay'] = delay_causes_airport.sum(axis=1)

# Select top 10 airports by total delays
top_airports = delay_causes_airport.nlargest(10, 'total_delay')

top_airports = top_airports.drop(columns='total_delay')

# Plot stacked bar chart
top_airports.plot(kind='bar', stacked=True, figsize=(14,6))
plt.title("Top 10 Airports with Most Delays (by Cause)", fontsize=14)
plt.xlabel("Airport")
plt.ylabel("Total Delay Minutes")
plt.legend(title="Cause of Delay")
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
# Group by airport and sum the delay causes
delay_causes_airport = new_df.groupby('airport')[['carrier_delay','weather_delay','nas_delay','security_delay','late_aircraft_delay']].sum()

# Calculate total delays per airport
delay_causes_airport['total_delay'] = delay_causes_airport.sum(axis=1)

# Select top 10 airports by total delays
top_airports = delay_causes_airport.nsmallest(10, 'total_delay')

top_airports = top_airports.drop(columns='total_delay')

# Plot stacked bar chart
top_airports.plot(kind='bar', stacked=True, figsize=(14,6))
plt.title("Top 10 Airports with Least Delays (by Cause)", fontsize=14)
plt.xlabel("Airport")
plt.ylabel("Total Delay Minutes")
plt.legend(title="Cause of Delay")
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
# Group by airline and sum the delay causes
delay_causes_airline = new_df.groupby('carrier_name')[['carrier_delay','weather_delay','nas_delay','security_delay','late_aircraft_delay']].sum()

# Calculate total delays per airline
delay_causes_airline['total_delay'] = delay_causes_airline.sum(axis=1)

# Select top 10 airlines by total delays
top_airlines = delay_causes_airline.nlargest(10, 'total_delay')

top_airlines = top_airlines.drop(columns='total_delay')

# Plot stacked bar chart
top_airlines.plot(kind='bar', stacked=True, figsize=(14,6))
plt.title("Top 10 Airlines with Most Delays (by Cause)", fontsize=14)
plt.xlabel("Airlines")
plt.ylabel("Total Delay Minutes")
plt.legend(title="Cause of Delay")
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

These charts highlights the airports and airlines with the highest total delays and breaks down the main causes. For example, we can see that at some airports, delays are often cause by late aircraft arrivals, while at others, weather or carrier-related issues take the lead. This tells us that not all airports struggle with delays for the same reasons. This can be helpful when looking for solutions to help each airport's challenges. Listed above, I have a visual of all the carriers and the breakdown of delay causes, however, I also wanted to highlight the top 5 airlines as well to go more in depth. Furthermore, we can see that most of the causes were due to late aircraft arrivals more than any other factors. This can be due to the high volume of planes within the company, or other circumstances.
